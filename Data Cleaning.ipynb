# Data Importation and Cleaning

```python
import pandas as pd
```

```python
df = pd.read_csv('Channel_attribution.csv')
df = df.fillna(0)
```

```python
df.head()
```

```python
cols = df.columns
```

Change all ints to str and remove any trailing .0s

```python
for col in cols:
    df[col] = df[col].astype(str)
    df[col] = df[col].map(lambda x : str(x)[:-2] if '.' in x else str(x))
    df[col] = df[col].map(lambda x : x.replace('0', ''))
```

```python
df.head()
```

Create path feature

```python
df['Path'] = ''
for i in df.index:
    for x in cols:
        df.at[i, 'Path'] = df.at[i, 'Path'] + df.at[i, x] + ' '
```

```python
df['Path'].head()
```

Split path on conversion

```python
df['Path'] = df['Path'].map(lambda x: x.split('21')[0])

df['Conversion'] = 1
```

Take just the paths and conversions

```python
df = df[['Path', 'Conversion']]
```

```python
df
```

```python
df['Path'] = df['Path'].apply(lambda x: x.replace(' ', ' >')[:-1])
```

```python
df
```

```python
df['Path'].value_counts()
```

```python
df = df.groupby('Path').sum().reset_index()
```
